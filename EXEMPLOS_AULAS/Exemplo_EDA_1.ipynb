```python
#importando a lib pandas
import pandas as pd
```

```python
dados = pd.read_csv('https://archive.ics.uci.edu/static/public/53/data.csv')
```

```python
#algumas operações sobre meu dataframe 'dados'
#3 primeiras linhas da tabela, por exemplo
dados.head(3)
```

```python
dados.tail(3)
```

```python
# prompt: apresente as colunas de um dataframe chamado dados

dados.columns
```

```python
# prompt: para um df chamado dados com as colunas: "['sepal length', 'sepal width', 'petal length', 'petal width', 'class']", renomei as colunas para nomes mais intuitivos em pt-br, mude a coluna class para categoria

dados = dados.rename(columns={
    'sepal length': 'comprimento_sepala',
    'sepal width': 'largura_sepala',
    'petal length': 'comprimento_petala',
    'petal width': 'largura_petala',
    'class': 'categoria'
})
```

```python
dados.columns.to_list()
```

# um pouco de estatística descritiva

```python
dados['largura_sepala'].max()
```

```python
dados['largura_sepala'].min()
```

```python
dados['largura_sepala'].mean()
```

```python
dados['largura_sepala'].median()
```

```python
dados['largura_sepala'].std()
```

```python
dados
```

```python
for coluna in dados.columns:
    if coluna == 'categoria':
        continue
    print(coluna)
    print(f'max da coluna= {dados[coluna].max()}')
    print(f'min da coluna= {dados[coluna].min()}')
    print(f'mean da coluna= {dados[coluna].mean()}')
    print(f'median da coluna= {dados[coluna].median()}')
    print(f'std da coluna= {dados[coluna].std()}')
```
